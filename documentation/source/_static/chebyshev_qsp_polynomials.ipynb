```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial import Chebyshev
from scipy.special import erf
```

## Plot inversion polynomial

```python
from qrisp.algorithms.gqsp.inversion import _inversion_cheb

theta = 0.2
delta = theta / 2
cheb = Chebyshev(_inversion_cheb(theta, 0.01))
print("degree cheb", cheb.degree())

def f(x):
    return 1.0 / x

# Evaluate for plotting
x_vals = np.linspace(-1, 1, 1000)
x_vals_l = np.linspace(-1, -theta + delta, 1000)
x_vals_r = np.linspace(theta - delta, 1, 1000)
y_cheb = cheb(x_vals)
y_inv_l = f(x_vals_l)
y_inv_r = f(x_vals_r)

plt.figure(figsize=(10, 6))
plt.axvline(x=theta, color='#20306f', linestyle='--')
plt.axvline(x=-theta, color='#20306f', linestyle='--')
plt.plot(x_vals_l, y_inv_l, 'r--', linewidth=2, color='#444444')
plt.plot(x_vals_r, y_inv_r, 'r--', linewidth=2, color='#444444')
plt.plot(x_vals, y_cheb, 'r--', linewidth=4, color='#6929C4')

plt.ylim(-10, 10)
plt.title("Inversion polynomial", fontsize=18, color="#444444" )
plt.xlabel("x", fontsize=18, color="#444444")
plt.ylabel("f(x)", fontsize=18, color="#444444")
plt.legend(fontsize=12, labelcolor="#444444")
plt.tick_params(axis="both", labelsize=12)
plt.grid()
plt.tight_layout()
plt.savefig("chebyshev_inversion.png")
plt.show()
```

## Plot pseudo-inversion polynomial

```python
from qrisp.algorithms.gqsp.pseudo_inversion import _pseudo_inversion_cheb

theta = 0.4
delta = theta / 4
cheb = Chebyshev(_pseudo_inversion_cheb(theta, delta, 0.01))
print("degree cheb", cheb.degree())

def f(x):
    return 1.0 / x

# Evaluate for plotting
x_vals = np.linspace(-1, 1, 1000)
x_vals_l = np.linspace(-1, -theta+delta, 1000)
x_vals_r = np.linspace(theta-delta, 1, 1000)
y_cheb = cheb(x_vals)
y_inv_l = f(x_vals_l)
y_inv_r = f(x_vals_r)

plt.figure(figsize=(10, 6))
plt.axvline(x=theta, color='#20306f', linestyle='--')
plt.axvline(x=theta - 2 * delta, color='#20306f', linestyle='--')
plt.axvline(x=-theta, color='#20306f', linestyle='--')
plt.axvline(x=-theta + 2 * delta, color='#20306f', linestyle='--')
plt.plot(x_vals_l, y_inv_l, 'r--', linewidth=2, color='#444444')
plt.plot(x_vals_r, y_inv_r, 'r--', linewidth=2, color='#444444')
plt.plot(x_vals, y_cheb, 'r--', linewidth=4, color='#6929C4')

plt.ylim(-3, 3)
plt.title("Pseudo-inversion polynomial", fontsize=18, color="#444444" )
plt.xlabel("x", fontsize=18, color="#444444")
plt.ylabel("f(x)", fontsize=18, color="#444444")
plt.legend(fontsize=12, labelcolor="#444444")
plt.tick_params(axis="both", labelsize=12)
plt.grid()
plt.tight_layout()
plt.savefig("chebyshev_pseudo_inversion.png")
plt.show()
```

## Plot kernel-reflection polynomial
